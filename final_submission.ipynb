2. David Hammer. 2022. Behind the Key Decision That Left Many Poor Homeowners Without Enough Money to Rebuild After Katrina. ProPublica. Retrieved April 11, 2024 from https://www.propublica.org/article/why-louisiana-road-home-program-based-grants-on-home-values
  * Through David Hammer’s *Behind the Key Decision That Left Many Poor Homeowners Without Enough Money to Rebuild After Katrina*, we can learn more about the political negotiations and decision-making processes that influenced aid distribution after Hurricane Katrina. This investigative piece sheds light the Road Home Program calculated rebuilding grants based on the pre-storm market values of homes rather than the actual costs of repairs needed. The formula for aid was as follows: Grant Amount = (Lower of Pre-Storm Home Value or Damage Assessment) - Insurance Payments - FEMA Aid - Penalties for Lack of Insurance. The maximum possible grant was capped at $150,000. This approach systematically disadvantaged homeowners in poorer neighborhoods where market values were lower than the actual rebuilding costs. Because of this equation, people in the poorest areas of New Orleans had to cover 30% of their rebuilding costs after aid and insurance. In the wealthiest areas, where residents had far more resources the shortfall was 20%. The decision to use pre-storm home values was influenced by concerns about federal spending and potential misuse of funds due to perceived corruption in Louisiana. Negotiations between Louisiana officials and the federal government led to a compromise where home values were used as a cost-controlling measure. The use of home values to determine aid amounts was unique to the situation following Hurricanes Katrina and Rita and has not been allowed in subsequent disaster responses. The U.S. Department of Housing and Urban Development (HUD) later required that aid be based on verified repair expenses rather than home values. The article is valuable because it is thorough in tracing the decision-making process behind the use of home values, providing a clear narrative of political and economic pressures that shaped the policy. It also highlights the real-world impact on affected homeowners, offering a critical look at the consequences of policy decisions. However, the narrative strongly emphasizes the failures of the program, which is important for accountability but might lead to an imbalance by not sufficiently acknowledging any successes or the complexities faced by policymakers in a post-disaster context. A more balanced view could help in understanding the full spectrum of challenges and achievements of the Road Home Program. As our research aims to observe the relationship between aid distribution in areas and correlations with economic growth or demographic shifts, we hope to discover findings that reveal how pre-storm market values may have understimated the true rebuilding costs in poorer areas. With the decision to based aid on pre-storm values, we can evaluate to what extent poorer homes were disadvantaged.

  * Ma and Smith's (2020) study, *Vulnerability of Renters and Low-Income Households to Storm Damage: Evidence from Hurricane Maria in Puerto Rico*, delves into the disproportionate impact of natural disasters on marginalized communities. They found that renters and low-income households are more likely to have their homes destroyed or majorly damaged. Despite constituting 8% of those experiencing severe damages, renters comprised a staggering 66% of households with severe damage. Renter-occupied homes were also four times more likely to be completely destroyed compared to owner-occupied ones. This aligns with the study's income analysis, revealing that lower-income households faced significantly higher risks of total destruction. The authors support their claims by drawing parallels with earlier research, and they cite Kamel (2019), who found a correlation between zip code areas with higher low-income populations and greater housing damage after Hurricane Katrina. Similarly, Logan (2005) reported that neighborhoods with higher renter concentrations saw increased damage during the same hurricane. Ma and Smith acknowledge the difficulty in definitively disentangling the independent effects of housing tenure and income on damage severity, but they argue that the two factors are inherently intertwined, making a clear separation potentially unhelpful. The study presents limitations beyond the inherent complexities of the research question, like the use of a fixed dollar threshold for damage classification ("minor," "major," and "destroyed") raises concerns about the potential for misrepresenting relative damage compared to home value. Additionally, observer bias might skew judgments on what constitutes "repairable" damage. Furthermore, the authors raise ethical concerns about FEMA's housing assistance program, which requires "stable housing" pre-disaster. This requirement could disadvantage already vulnerable low-income renters struggling with housing insecurity, potentially leaving them with even less recovery support. As Ma & Smith linke housing tenure to socioeconomic factors like income, our research will similarly analyze how income disparities may have affected displacement due to insufficient repair funds. If our work reveals that neighborhoods with lower aid ratios have a higher percentage of residents with lower income levels, this will further validate Ma & Smith's argument that lower income can increase vulnerability to storm damage and hinder recovery efforts.

## **Exploring the Distribution of Closing Damage Ratio (Amount of Damage/Aid Received)**

```python
aggregated_data = new_orleans_data.groupby('NOLA Neighborhood Name').agg({
    'Closing Damage Assessment': 'sum',
    'TOTAL_CLOSING_AMOUNT': 'sum',

}).sort_values(by='Closing Damage Assessment', ascending=False)

aggregated_data_reset = aggregated_data.reset_index()


# Convert 'Current Damage Assessment' and 'TOTAL_CLOSING_AMOUNT' to millions for plotting
aggregated_data_reset['Closing Damage Assessment (Millions)'] = aggregated_data_reset['Closing Damage Assessment'] / 1e6
aggregated_data_reset['TOTAL_CLOSING_AMOUNT (Millions)'] = aggregated_data_reset['TOTAL_CLOSING_AMOUNT'] / 1e6

# Plotting
fig, ax = plt.subplots(figsize=(14, 8))

# Set the positions and width for the bars
positions = np.arange(len(aggregated_data_reset['NOLA Neighborhood Name']))
width = 0.35  # the width of the bars

# Plot bars for total damage assessment and total aid amount in millions
plt.bar(positions - width/2,
        aggregated_data_reset['Closing Damage Assessment (Millions)'],
        width,
        alpha=0.7,
        color='red',
        label='Total Damage Assessment (Millions)')

plt.bar(positions + width/2,
        aggregated_data_reset['TOTAL_CLOSING_AMOUNT (Millions)'],
        width,
        alpha=0.7,
        color='green',
        label='Total Aid Amount (Millions)')

# Adding some additional features to the plot
ax.set_ylabel('Amount (Millions)')
ax.set_title('Total Damage Assessment and Aid Amount by Neighborhood (in Millions)')
ax.set_xticks(positions)
ax.set_xticklabels(aggregated_data_reset['NOLA Neighborhood Name'], rotation=90)
plt.legend(loc='upper right')
plt.tight_layout()

plt.show()
```

The bar graph above juxtaposes the estimated damages and the aid allocated to each neighborhood in New Orleans, which is pivotal for revealing any substantial discrepancies between the two. These discrepancies, especially in neighborhoods that are identified as marginalized, could provide evidence supporting the hypothesis that aid distribution was not only uneven but also detrimentally affected these vulnerable communities. Additionally, the magnitude of difference between the total amount damage assessed and aid given across all neighborhoods is one of concern as it hints the need for an investigation into the mechanisms of aid distribution. The chart aligns well with our concern in our hypothesis, as it continues to allow us to question whether the assistance provided truly aligned with the actual level of need in each community, or whether systemic biases skewed the distribution, thereby perpetuating a cycle of disadvantage for the marginalized areas.



We can also see that no neighbourhood on average got enough aid, since the mean, median, low, and max are all above 0 for the damage/aid ratio:

```python
mean_ratio = demographic_data['Closing Damage Assessment/Closing Amount'].mean()
median_ratio = demographic_data['Closing Damage Assessment/Closing Amount'].median()
min_ratio = min(demographic_data['Closing Damage Assessment/Closing Amount'])
max_ratio = max(demographic_data['Closing Damage Assessment/Closing Amount'])

print("Mean ratio: " + str(mean_ratio.round(3)))
print("Median ratio: " + str(median_ratio.round(3)))
print("Minimum ratio: " + str(round(min_ratio, 3)))
print("Maximum ratio: " + str(round(max_ratio, 3)))
```

### *Race*

```python
# Create a new df with relevant details: NOLA Neighborhood Name, colored_neighborhood classification, (Closing Damage Assessment/Closing Amount)

#create a new column for colored neighborhoods, if a community has 49% or less white population they are considered colored
df['POC majority'] = df['White Population 2000'] <= .49

new_orleans_data = pd.read_csv('new_orleans_data.csv')
#merge data
race_aid_analysis = pd.merge(new_orleans_data[['NOLA Neighborhood Name', 'Closing Damage Assessment/Closing Amount']],
                             df[['Neighborhood_name', 'POC majority']],
                             left_on='NOLA Neighborhood Name',
                             right_on='Neighborhood_name',
                             how='left')


clean_race_aid = race_aid_analysis.dropna()
```

```python
# Separate the neighborhoods into colored and non-colored based on the binary indicator
colored_neigh_aid = clean_race_aid[clean_race_aid['POC majority']]['Closing Damage Assessment/Closing Amount']
non_colored_neigh_aid = clean_race_aid.loc[clean_race_aid['POC majority'] == False, 'Closing Damage Assessment/Closing Amount']

# Get summary statistics for colored neighborhoods
colored_neigh_summary = clean_race_aid[clean_race_aid['POC majority']]['Closing Damage Assessment/Closing Amount'].describe()

# Get summary statistics for non-colored neighborhoods
non_colored_neigh_summary = clean_race_aid.loc[clean_race_aid['POC majority'] == False]['Closing Damage Assessment/Closing Amount'].describe()

# Perform a two-sample t-test to see if there is a significant difference in the means
t_stat, p_value = stats.ttest_ind(colored_neigh_aid, non_colored_neigh_aid, equal_var=False)

print("T Stat:", t_stat)
print("P Value:",p_value)
print("-----------------------------------")
print("Summary for POC majority Neighborhoods:")
print(colored_neigh_summary)
print("\nSummary for white majority Neighborhoods:")
print(non_colored_neigh_summary)
```

```python
# Create a new column in 'clean_race_aid' to denote the neighborhood type
clean_race_aid['Neighborhood Type'] = clean_race_aid['POC majority'].apply(lambda x: 'POC majority' if x else 'white majority')

# Plotting
plt.figure(figsize=(12, 6))

# Box Plot
plt.subplot(1, 2, 1)
sns.boxplot(x='Neighborhood Type', y='Closing Damage Assessment/Closing Amount', data=clean_race_aid)
plt.title('Box Plot of Closing Damage Assessment/Closing Amount')
plt.yscale('log') # Using a log scale due to wide distribution ranges


plt.tight_layout()
plt.show()
```

With a p-value of approximately 0.023, the result is considered statistically significant at the 0.05 level, suggesting that the difference in means between the two groups (presumably, 'colored' vs. 'non-colored' neighborhoods in terms of aid distribution) is unlikely to have occurred by random chance alone. This statistical significance supports the hypothesis that aid distribution has been uneven, which could have ramifications for marginalized communities.

POC majority neighborhoods have a higher mean (3.95 for POC and 3.23 for white)and much higher maximum Closing Damage Assessment/Closing Amount ratio than white majority neighborhoods, indicating that on average, and especially in extreme cases (indicated by outliers), POC neighborhoods received less aid relative to the damage assessed. This is highlighted by the standard deviation and the maximum values in the POC majority neighborhoods, which are significantly larger than those in white majority neighborhoods, revealing that POC majority neighborhoods experienced a much wider variance in the ratio (44.43 std for POC and 10.35 for white), with some areas receiving a disproportionate amount of aid compared to the damages assessed.

The attached box plot visually reinforces these findings. The spread of the data points for POC majority neighborhoods is wider, with more extreme outliers, suggesting that while the median is somewhat similar across both types of neighborhoods, POC majority neighborhoods experienced more instances of extreme ratios, where the aid did not correspond as closely to the damages assessed.

### *Income Levels*

```python
# Define the income levels as low, medium, high based on quantiles
df['income_level'] = pd.qcut(df['avg_household_income_2000_inflated'], 3, labels=['low', 'medium', 'high'])

# Create a new df with relevant details: NOLA Neighborhood Name, low education classification, (Closing Damage Assessment/Closing Amount)
income_aid_analysis = pd.merge(new_orleans_data[['NOLA Neighborhood Name', 'Closing Damage Assessment/Closing Amount']],
                             df[['Neighborhood_name', 'income_level']],
                             left_on='NOLA Neighborhood Name',
                             right_on='Neighborhood_name',
                             how='left')

#drop na values
clean_income_aid = income_aid_analysis.dropna()

low_income_aid = clean_income_aid[clean_income_aid['income_level'] == 'low']['Closing Damage Assessment/Closing Amount']
med_income_aid = clean_income_aid[clean_income_aid['income_level'] == 'medium']['Closing Damage Assessment/Closing Amount']
high_income_aid = clean_income_aid[clean_income_aid['income_level'] == 'high']['Closing Damage Assessment/Closing Amount']

low_income_summary = low_income_aid.describe()
med_income_summary = med_income_aid.describe()
high_income_summary = high_income_aid.describe()

# Summary statistics and visualization
print("Summary Statistics by Income Level:")
print(clean_income_aid.groupby('income_level')['Closing Damage Assessment/Closing Amount'].describe())

# ANOVA to test for differences in means across the three income levels
anova_result = stats.f_oneway(clean_income_aid[clean_income_aid['income_level'] == 'low']['Closing Damage Assessment/Closing Amount'],
                              clean_income_aid[clean_income_aid['income_level'] == 'medium']['Closing Damage Assessment/Closing Amount'],
                              clean_income_aid[clean_income_aid['income_level'] == 'high']['Closing Damage Assessment/Closing Amount'])

print("\nANOVA Test Results:")
print("F-statistic:", anova_result.statistic)
print("P-value:", anova_result.pvalue)

# Visualization with boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='income_level', y='Closing Damage Assessment/Closing Amount', data=clean_income_aid)
plt.title('Closing Damage Assessment/Closing Amount by Income Level')
plt.show()

```

The p-value (0.0012) is very low, far below the conventional threshold of 0.05, indicating that the differences in mean aid ratios between the three income levels are statistically significant. This suggests that the level of income in a neighborhood is a significant factor in the distribution of aid, with different income levels receiving different amounts of aid relative to damage assessments. The findings imply that the distribution of aid was not uniform across income levels and that middle-income neighborhoods, on average, received a higher aid ratio than both low and high-income neighborhoods. It also raises questions about the equity of aid distribution and whether it effectively addressed the needs based on the damages suffered.

On the other hand, the highest mean aid ratio in the medium-income group could indicate that these neighborhoods received a larger proportion of aid relative to damage, or it could be influenced by outliers where a few neighborhoods received very large amounts of aid. The lower mean aid ratio in the high-income group compared to the medium-income group may suggest that these neighborhoods had less damage, more resources to recover without aid, or that the aid was more evenly distributed without the influence of extreme outliers.

These findings bolster our hypothesis that aid was not equitably allocated. The notable difference in aid ratios between income groups, highlights potential biases in the aid distribution process. These biases could have disproportionately favored certain income brackets, supporting the hypothesis that theprogram may have contributed further to existing socioeconomic disparities rather than closing them.

### *Relationship between close damage ratio and the demographics of neighborhoods*

```python
def calculate_correlation(column1, column2):
    return df['Closing Damage Assessment/Closing Amount'].corr(column1), df['Closing Damage Assessment/Closing Amount'].corr(column2)

def print_correlation_stats(population, column_2000, column_2020):
    population_change = column_2020 - column_2000
    mean_change = np.mean(population_change)
    print(f"{population} population")
    print(f"There's a {round(mean_change, 2)}% change in {population.lower()} populations")
    correlation_ratio_vs_population_change = df['Closing Damage Assessment/Closing Amount'].corr(population_change)
    print("Correlation between damage/aid ratio and: ")
    print(f"    the increase of {population.lower()} populations in all neighborhoods: {round(correlation_ratio_vs_population_change, 2)}")
    correlation1, correlation2 = calculate_correlation(column_2000, column_2020)
    print(f"    the percent of {population.lower()} residents in a neighborhood before Katrina: {round(correlation1, 2)}")
    print(f"    the percent of {population.lower()} residents in a neighborhood after Katrina: {round(correlation2, 2)}")
    print()


print_correlation_stats("Black", df['Black Population 2020'], df['Black Population 2000'])
print_correlation_stats("Indigenous", df['Indigenous Population 2000'], df['Indigenous Population 2020'])
print_correlation_stats("Asian", df['Asian Population 2000'], df['Asian Population 2020'])
print_correlation_stats("Hispanic", df['Hispanic Population 2000'], df['Hispanic Population 2020'])
print_correlation_stats("White", df['White Population 2000'], df['White Population 2020'])
```

We also wanted to explore how demographic data over time changed after Hurricane Katrina to see how communities recovered. We used to have education as a plot but removed it due to mixed results.

```python
def calculate_correlation(column1, column2):
    change_in_demographics = column2 - column1
    return df['Closing Damage Assessment/Closing Amount'].corr((change_in_demographics))

def get_correlation_diff(population, column_2000, column_2020):
  correlation = calculate_correlation(column_2000, column_2020)
  return correlation

# Data
populations = ["Black", "Indigenous", "Asian", "Hispanic", "White"]
correlation_diffs = [get_correlation_diff(population,
                                          df[f'{population} Population 2000'],
                                          df[f'{population} Population 2020']) for population in populations]

# Plot
plt.figure(figsize=(10, 6))
plt.barh(populations, correlation_diffs, color='skyblue')
plt.xlabel('Correlation Difference Before and After Katrina')
plt.ylabel('Population')
plt.title('Correlation Between the Damage/Aid ratio of a neighborhood and the change in demographics')
plt.axvline(0, color='grey', linewidth=0.5)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

We can see that there is a statistically significant difference in the change of Black and white people in each neighborhood correlated with the amount of damage/aid ratio that neighborhood got (as a reminder, the higher the damage/aid ratio is, the worse off that area was).

This means that if the damage/aid ratio got higher for a neighborhood, it was more likely to have its Black population grow from 2000 to 2020. Whereas areas that had a lower damage/aid ratio, the white population grew from 2000 to 2020. Other demographics did not have a statistically significant changes in demographics.



```python

# Calculate income change and white population change percentages
df['Income Grew Percentage'] = ((df['avg_household_income_2020'] - df['avg_household_income_2000_inflated']) / df['avg_household_income_2000_inflated']) * 100
df['White Population Change Percentage'] = (df['White Population 2020'] - df['White Population 2000'])
df['Black Population Change Percentage'] = ((df['Black Population 2020'] - df['Black Population 2000']) / df['Black Population 2000'])

income_changed= []
white_pop_grew = []
gentrified = []
for index, row in df.iterrows():
  income_grew_40_percent = row['avg_household_income_2020']/row['avg_household_income_2000_inflated'] > 1.4
  white_population_grew = row['White Population Change Percentage'] >= 0.20
  if income_grew_40_percent and white_population_grew:
    gentrified.append(row['Neighborhood_name'])
  if white_population_grew:
    white_pop_grew.append(row['Neighborhood_name'])
  if income_grew_40_percent:
    income_changed.append(row['Neighborhood_name'])


# Filter data for gentrified neighborhoods
income_changed = df[df['Neighborhood_name'].isin(income_changed)]
white_pop_grew2 = df[df['Neighborhood_name'].isin(white_pop_grew)]
gentrified2 = df[df['Neighborhood_name'].isin(gentrified)]

# Analyze correlation between Closing Damage Assessment/Closing Amount and gentrification
correlation = df['Closing Damage Assessment/Closing Amount'].corr(df['Neighborhood_name'].isin(gentrified))
correlation2 = df['Closing Damage Assessment/Closing Amount'].corr(df['Income Grew Percentage'])
correlation3 = df['Closing Damage Assessment/Closing Amount'].corr(df['Income Grew Percentage'])

print("Income grew in neighborhoods:")
print(income_changed['Neighborhood_name'])
print()

print("Gentrified neighborhoods:")
print(gentrified)
print()

print("White population grew neighborhoods:")
print(white_pop_grew2['Neighborhood_name'])
print()

print("\nCorrelation between Closing Damage Assessment/Closing Amount and gentrification:")
print(correlation)

print("\nCorrelation between Closing Damage Assessment/Closing Amount and Income Grew Percentage:")
print(correlation2)
```

From this we can see that there's a strong negative correlation between the damage/aid ratio a neighborhood got and the economic growth of the population measured by the average household income in each area. There's also a moderate negative correlation between gentrification and the damage/aid ratio.

We considered an area “gentrified” if the white population grew at least 20 percent and if the area’s average household income grew at least 40 percent in a neighborhood from 2000 to 2020.
* First, we analyzed if neighborhoods’ changes in demographics had any correlation to the damage/aid ratio of that area. We found that there was a statistically significant difference in the change of Black and white people in each neighborhood correlated with the amount of damage/aid ratio that neighborhood got (as a reminder, the higher the damage/aid ratio is, the worse off that area was). This means that if the damage/aid ratio got higher for a neighborhood, it was more likely to have its Black population grow from 2000 to 2020. Whereas areas that had a lower damage/aid ratio, the white population grew from 2000 to 2020. Other demographics did not have statistically significant changes in demographics.
This could mean the following:
  * Higher damage might deter investment: It's possible that areas with very high damage/aid ratios have suffered significant blight or infrastructure issues that outweigh potential benefits for investors, even with cheaper property prices.
  * Gentrification prioritizes areas with existing resources: Wealthy white people and/or companies might be more likely to invest in areas that already have some level of infrastructure and amenities, even if those areas have historically received less aid. This could be due to factors like lower renovation costs, existing transportation networks, or proximity to “desirable” locations for wealthy populations.
* We found a slightly negative correlation between the gentrified neighborhood and the damage/aid ratio, meaning that the higher the damage/aid ratio was for a neighborhood, the less likely it was going to be gentrified, although this correlation was only -0.19. This was not what we originally expected, since we predicted that the neighborhoods that had higher damage/aid ratios would be more likely that wealthy white people and/or companies would buy cheap property here, renovate, and sell to a higher income population. This seems to be untrue from our correlation outputs.
* Next, we wanted to see if the damage/aid ratio had any effects in the economic growth of a neighborhood. Comparing the average household income from 2000 to the average household income from 2020, we saw that there's a strong negative correlation between the damage/aid ratio a neighborhood got and the economic growth of the population measured by the average household income in each area. This was a crucial and important finding, because this means that an algorithmic decision that decided how much aid the average household in a neighborhood got significantly impacted the neighborhood’s economic growth in the next 15-20 years.

Our research tested the hypothesis that the Road Home Program's aid distribution was uneven and adversely affected marginalized communities, thereby exacerbating socioeconomic disparities and contributing to gentrification and displacement. Our analysis supported this, showing a statistically significant difference in aid allocation between marginalized and affluent white neighborhoods, indicating that the disparity was not due to random chance.

**Data Limitations**

When using the dataset related to the Road Home program, an important limitation that we want to highlight is that the dataset may not encapsulate the entire population of those that were affected by the hurricane. Specifically, the data may not account for people who never applied for or were unaware of the availability of these hurricane relief aid. This can result in a biased understanding of the aid’s distribution and impact as it could be omitting details about a group that may have needed funding but did/could not access the program due to lack of information or other barriers.

**Model Limitations**
We also made assumptions for determining the threshold for low-income communities by performing a qcut of the entire dataset. Thus, income levels for our analysis is only based on the data from the Road Home Program and may not be fully representative of the population.

**Analysis Limitations**

We defined gentrification based on two main indicators: changes in racial demographics over time and significant income growth, specifically a 40% increase over the past decade. While there is no universally accepted formula for identifying gentrification, our criteria were informed by various online resources and the data available to us. These two variables were selected as they effectively capture the socio-economic and demographic shifts typically associated with gentrification.
