```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read the different csv
dataset = pd.read_csv('roadhome_for_datastore.csv')
race_by_neighborhood_2000 = pd.read_csv('race_by_neighborhood_2000.csv')
race_by_neighborhood_2020 = pd.read_csv('race_by_neighborhood_2020.csv')
education_level = pd.read_csv('education_level.csv')
poverty_level = pd.read_csv('poverty_percentage.csv')
avg_household_income = pd.read_csv('avg_household_income.csv')
rent_change = pd.read_csv('rent_changes.csv')
```

```python
# Filter and group the data for New Orleans neighborhoods
f_data = dataset[dataset["Closing Option"] == 1]
new_orleans_data = f_data[f_data['GIS City'] == "NEW ORLEANS"]

# Remove rows with "NOLA Neighborhood Name" equal to "Iberville Development"
new_orleans_data = new_orleans_data[new_orleans_data["NOLA Neighborhood Name"] != "Iberville Development"]
new_orleans_data = new_orleans_data[new_orleans_data["NOLA Neighborhood Name"] != "East Carrollton"]
new_orleans_data = new_orleans_data[new_orleans_data["NOLA Neighborhood Name"] != "Florida Development"]
new_orleans_data = new_orleans_data[new_orleans_data["NOLA Neighborhood Name"] != "Desire Area"]

nola_groups = new_orleans_data.groupby("NOLA Neighborhood Name")

# Calculate the mean difference for each neighborhood
new_orleans_data['Mean_Difference'] = round(new_orleans_data['TOTAL_CLOSING_AMOUNT'] - new_orleans_data['Current Damage Assessment'])
mean_difference = nola_groups.agg({'Mean_Difference': 'mean'})

new_orleans_data['Closing Damage Assessment/Closing Amount'] = round(new_orleans_data['Closing Damage Assessment']/new_orleans_data['TOTAL_CLOSING_AMOUNT'])
mean_difference_of_ratio = nola_groups.agg({'Closing Damage Assessment/Closing Amount': 'mean'})

mean_difference
output_table = pd.DataFrame({
    'Neighborhood_name': race_by_neighborhood_2000['Neighborhood'],
    'Mean Difference': mean_difference['Mean_Difference'].values,
    'Closing Damage Assessment/Closing Amount': mean_difference_of_ratio['Closing Damage Assessment/Closing Amount'].values,
    'Black Population 2000':race_by_neighborhood_2000['Black_2000'],
    'Black Population 2020':race_by_neighborhood_2020['Black_2020'],
    'White Population 2000':race_by_neighborhood_2000['White_2000'],
    'White Population 2020':race_by_neighborhood_2020['White_2020'],
    'Asian Population 2000':race_by_neighborhood_2000['Asian_2000'],
    'Asian Population 2020':race_by_neighborhood_2020['Asian_2020'],
    'Indigenous Population 2000':race_by_neighborhood_2000['Indigenous_200'],
    'Indigenous Population 2020':race_by_neighborhood_2020['Indigenous_2020'],
    'Multiracial Population 2000':race_by_neighborhood_2000['Multiracial_2000'],
    'Multiracial Population 2020':race_by_neighborhood_2020['Multiracial_2020'],
    'Hispanic Population 2000':race_by_neighborhood_2000['Hispanic_2000'],
    'Hispanic Population 2020':race_by_neighborhood_2020['Hispanic_2020'],
    'Other Population 2000': race_by_neighborhood_2000['Other_2000'],
    'Other Population 2020':race_by_neighborhood_2020['Other_2020']
})

output_table

```

```python
#create a demographic data table
pov_educ_demographic_data = pd.merge(poverty_level, education_level, left_on='Unnamed: 0', right_on='neighborhood', how='inner')

# merging income and rent change with the table as well
avg_income_data = pd.merge(avg_household_income,rent_change,  left_on='neighborhood', right_on='neighborhood', how='inner')

merged = pd.merge(pov_educ_demographic_data, avg_income_data)

merged_demographic_data = pd.merge(output_table, merged, left_on='Neighborhood_name', right_on='neighborhood', how='inner')

merged_demographic_data = merged_demographic_data.drop(columns=['Unnamed: 0'])

merged_demographic_data = merged_demographic_data.fillna(0)

merged_demographic_data
```

```python
string_format = ['Neighborhood_name', 'Closing Damage Assessment/Closing Amount']

# Define the columns containing dollar and percentage values
dollar_columns = ["avg_household_income_2000_inflated",  "avg_household_income_2020",  "40th_percentile_2020", "67th_percentile", "contract_rent_2000_inflated", "contract_rent_2020", "gross_rent_2006_inflated", "gross_rent_2020_inflated"]
percentage_columns = [col for col in merged_demographic_data.columns if col not in dollar_columns and col not in ['Neighborhood_name', 'Closing Damage Assessment/Closing Amount']]
# Define the columns containing dollar and percentage values

for col in merged_demographic_data.columns:
  merged_demographic_data[col] = merged_demographic_data[col].astype(str)  # cast to string
  if col in dollar_columns:
    merged_demographic_data[col] = merged_demographic_data[col].str.replace('$', '')
    merged_demographic_data[col] = merged_demographic_data[col].str.replace(',', '')
    merged_demographic_data[col] = merged_demographic_data[col].astype(float)
  elif col not in ['Neighborhood_name', 'Closing Damage Assessment/Closing Amount']:  # Assuming these don't need conversion
    try:
      merged_demographic_data[col] = merged_demographic_data[col].str.replace('%', '').replace(',', '').astype(float) / 100
    except:
      pass
```

```python
updated_demographics = merged_demographic_data

merged_demographic_data.to_csv("updated_demographics3")
```
